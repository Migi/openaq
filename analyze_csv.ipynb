```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("openaq_data.csv")
df["utc"] =  pd.to_datetime(df["utc"], format='%Y-%m-%dT%H:%M:%S+00:00')
df
```

```python
df_hour_of_day = df.groupby(df["utc"].map(lambda x: str(x.hour).zfill(2) + ":00")).agg(
    count = ("value", "count")
)
df_hour_of_day
```

```python
%matplotlib inline

# Pie chart, where the slices will be ordered and plotted counter-clockwise:
labels = df_hour_of_day.index
sizes = df_hour_of_day["count"]

fig, ax = plt.subplots()
ax.pie(sizes, labels=labels, autopct='%1.1f%%')
ax.axis('equal')  # Equal aspect ratio ensures the pie chart is circular.
ax.set_title('Engineering Diciplines')

plt.show()
```

```python
time_df = df.groupby(pd.Grouper(key="utc", freq="1d")).agg(
    count = ("parameter", "count"),
    value_sum = ("value", "sum"),
    avg = ("value", "mean")
)
time_df
```
